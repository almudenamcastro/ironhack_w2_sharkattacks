```python
import re

def limpiar_fecha(source):
    if isinstance(source, str):  # Verifica si el valor es una cadena
        return re.sub(r', \d{1,2}/\d{1,2}/\d{4}', '', source)
    return source  # Retorna el valor sin cambios si no es una cadena

```

```python
shark_attack_row['source'] = shark_attack_row['source'].apply(limpiar_fecha)
```
